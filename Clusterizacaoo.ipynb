```python
stdscaler = StandardScaler()
# yHatSTD = previsto em cima de flowers padronizado.
flowers_std = stdscaler.fit_transform(flowers)
km = KMeans(n_clusters=3)
yHatSTD = km.fit_predict(flowers_std)
```

```python
# Só muda o tipo de normalização aplicada. Em um foi usado MinMaxScaler e no outro StandardScaler.
```

```python
dfIris['target_std'] = yHatSTD
```

```python
# Nesse caso conseguimos comparar os erros, pois já conhecemos as classes.
# Porém não é comum termos a classe em um problema de clusterização.
# Não existe certo ou errado em clusterização. Tentamos minimizar os erros, normalizando os dados, retirando os outliers e tentando tirar o máximo de interferências.
# Conversar com alguém do negócio, que vai analisar de outra maneira, descrevendo os grupos, para saber se estão diferentes, por exemplo.
# Ou aumentar o número de clusters, para verificar se há mais divisões. Onde essa divisão ocorre, sabemos que esse grupo é mais fraco.
```

```python
# Método do cotovelo para ver se 3 é um bom número de clusters do problema.
wcss = []
for i in range(1,10):
  km = KMeans(n_clusters=i)
  km.fit(flowers)
  wcss.append(km.inertia_)
```

```python
plt.plot(range(1,10),wcss)
plt.show()
```
